```python
# Chance of Admit Target
# Inputs to be in same scale use min-max scaler as max value is known
# 1 input 1 hidden and 1 output layers
# input 7, hidden 7 nodes and output 1
# 7x7+7 bias = 56 para / 7x1+1 bias = 8 - trainable parameters
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=40)
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = Sequential()
model.add(Dense(units=7,activation='relu',input_dim=7))
model.add(Dense(units=14,activation='relu'))
model.add(Dense(units=1,activation='linear'))
model.summary()
```

```python
model.compile(loss='mean_absolute_error',metrics=['mae'],optimizer='Adam')
```

```python
history = model.fit(X_train_scaled,y_train,epochs=1000,validation_split=0.2)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
r2_score(y_test,y_pred)
```

```python
loss_check = pd.DataFrame(history.history)
px.line(loss_check.rename_axis('Epoch'))
```
