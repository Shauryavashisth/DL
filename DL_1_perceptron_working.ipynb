```python
import pandas as pd
from sklearn.linear_model import Perceptron
from sklearn.datasets import make_classification
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv(r'D:\Customer-Churn-Records.csv')
df = df[['CreditScore','Age','Exited']].head(1000)#,'Tenure','Balance','EstimatedSalary','Exited']]
df.head()
```

```python
x= df.drop(columns='Exited')
y= df['Exited']
```

```python
p = Perceptron()
p.fit(x,y)
print('Coeff:-',p.coef_)
print('Intercept:-',p.intercept_)
```

### Example CreditScore - 619 and Age - 42
#### Ax+By+c
#### (619*-844)+(42*7821)+-42
if >0 = 1
else <0 = 0

```python
plot_decision_regions(x.values,y.values,clf=p,legend=2)
```

### With Stable Data

```python
x,y = make_classification(n_samples=1000,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=32,hypercube=False,class_sep=10)
```

```python
p = Perceptron()
p.fit(x,y)
print('Coeff:-',p.coef_)
print('Intercept:-',p.intercept_)
```

```python
plot_decision_regions(x,y,clf=p,legend=2)
```
